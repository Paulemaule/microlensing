```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
results_path = 'H:/Microlensing/data/'
if not os.path.exists(results_path): results_path = '/home/Tit5/paul.zuern/microlensing/data/'
```

```python
data = None
with open(results_path + 'out.txt', 'r') as f:
    keys = f.readline().split()
    collums = np.array([line.split() for line in f.readlines()]).astype(float).T
    data = dict(zip(keys, collums))
```

```python
data
```

```python
dist = np.linalg.norm(np.vstack([data['x1_ffp'],data['x2_ffp']]) - np.vstack([data['x1_star'],data['x2_star']]), axis = 0)
```
